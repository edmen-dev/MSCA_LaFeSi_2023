This notebook shows how to access and manipulate the most important data stored in this repository:
 1. `Weiss field` and `density of states` data are stored in marmot.hdf5
 2. The folder `calculation` contains all inputs and outputs, produced by `marmot code` (https://warwick.ac.uk/fac/sci/physics/research/theory/research/electrstr/marmot/)
 3. Metadata is described in `marmot.json` file.
 4. The folder `hdf5_and_metadata_class` contains a python class called by `set_hdf5.ipynb` used to create:
    - `marmot.hdf5`
    - `marmot.json`

# Reading hdf5 file

```python
f = h5py.File("marmot.hdf5", "r")
```

```python
# list of lattice parameters investigated
list_of_alat = [21.2735, 21.4906, 21.7077, 21.9247, 22.1418]

# name of the folder containing the calculations
folder0 = './calculation'
```

# Plots of central data

## Weiss fields vs magnetic order parameter

```python
for ilat, alat in enumerate(list_of_alat):
    current_folder_alat = folder0 + '/a' + str(alat)

    data_to_plot = f[current_folder_alat + '/weiss_field_vs_m'][:,:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], '-o', label='a = '+str(alat)+r' a$_0$')
    
plt.xlabel(r'order parameter $m$')
plt.ylabel(r'Weiss field (meV)')
plt.legend()
plt.grid()
```

## DOS vs magnetic order parameter

```python
for ilat, alat in enumerate(list_of_alat):
    current_folder_alat = folder0 + '/a' + str(alat)

    data_to_plot = f[current_folder_alat + '/dos_vs_m'][:,:]
    plt.plot(data_to_plot[:,0], data_to_plot[:,1], '-o', label='a = '+str(alat)+r' a$_0$')
    
plt.xlabel(r'order parameter $m$')
plt.ylabel(r'Density of states at the Fermi energy (1/eV)')
plt.legend(loc='lower left')
plt.grid()
```
